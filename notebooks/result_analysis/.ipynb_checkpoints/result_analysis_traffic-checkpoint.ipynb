## Scenario analysis

```python
x=experiments
y= outcomes['connectivity_mean'] < 36.635556
# y = (outcomes["num_of_nodes"] > 4000) & (outcomes["num_of_edges"] > 4000)
# z = 
prim_alg = prim.Prim(x, y, threshold=0.4)
box1 = prim_alg.find_box()
box1.show_tradeoff()
```
