```python
import pandas as pd
import osmnx as ox
import networkx as nx
import math
from numpy import linalg as LA
import numpy as np
```

```python
filepath="../graph/rotterdam_drive_bbox_cameras_traffic_lights_bridges_roundabouts_tunnels.graphml"
graph_test=ox.load_graphml(filepath)
```

```python
road_type = set()
for origin, destination, data_edge in graph_test.edges(data=True):
    for element in data_edge["highway"]:
     road_type.add(element)
```

```python
residential = 0
for origin, destination, data_edge in graph_test.edges(data=True):
    if data_edge["highway"] == "residential":
        residential +=1
```

```python
unclassified = 0
for origin, destination, data_edge in graph_test.edges(data=True):
    if data_edge["highway"] == "unclassified":
        unclassified +=1
```

```python
edge_color = []

for origin, destination, data_edge in graph_test.edges(data=True):
    if data_edge["highway"] == "residential":
        edge_color.append("red")
    elif data_edge["highway"] == "unclassified":
        edge_color.append("blue")
    else:
        edge_color.append("gray")
        


ox.plot.plot_graph(
            graph_test, bgcolor="white", node_color="lightgray", node_size=1, edge_linewidth=1,
            edge_color=edge_color
        )
```

```python
edge_color = []

for origin, destination, data_edge in graph_test.edges(data=True):
    if data_edge["highway"] == "residential":
        edge_color.append("gray")
    elif data_edge["highway"] == "unclassified":
        edge_color.append("gray")
    else:
        edge_color.append("red")
        


ox.plot.plot_graph(
            graph_test, bgcolor="white", node_color="lightgray", node_size=1, edge_linewidth=1,
            edge_color=edge_color
        )
```
