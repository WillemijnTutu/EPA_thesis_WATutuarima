```python
import numpy as np
import osmnx as ox
import os
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Notebook for the case study points

```python
default_graph_file_path = "../../graph/graph_base_case.graphml"
graph_OW_False = ox.load_graphml(default_graph_file_path)
```

```python
centre_point = 6238824713 #stadhuis 
highway_north = 44596978 
highway_east = 44471862 
residential_south = 44201093 # Emelissedijk met omloopdijk
```

```python
node_colors = []
node_size = []
edge_colors = []
edge_size = []
for index in graph_OW_False.nodes():
    if index == highway_east:
        node_colors.append('tab:blue')
        node_size.append(30)
    elif index == highway_north:
        node_colors.append('tab:blue')
        node_size.append(30)
    elif index == residential_south:
        node_colors.append('tab:blue')
        node_size.append(30)
    elif index == centre_point:
        node_colors.append('red')
        node_size.append(30)
    else:
        node_colors.append("lightgray")
        node_size.append(0)

for (u, v) in graph_OW_False.edges():
    edge_colors.append('lightgray')
    edge_size.append(0.5)

file_path = "case_study_points.png"


ox.plot.plot_graph(
    graph_OW_False,
    bgcolor="white", 
    node_color=node_colors, 
    node_size=node_size, 
    edge_linewidth=edge_size,
    edge_color=edge_colors,
    show=True, save=False, filepath=file_path
)
```
