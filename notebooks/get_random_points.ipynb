```python
import pandas as pd
import osmnx as ox
import networkx as nx
import math
import random
```

```python
filepath = "../graph/rotterdam_drive_with_cameras.graphml"
graph = ox.load_graphml(filepath)
```

```python
x = random.choice(list(graph.nodes()))
```

```python
y = random.choice(list(graph.nodes()))
```

```python
path = ox.distance.shortest_path(graph, x, y, weight='length')
```

```python
node_size = []
node_color = []
for node in graph.nodes:
    node_size.append(0)
    node_color.append('blue')


ox.plot.plot_graph_route(
                graph, list(path), route_color='b', route_linewidth=4, route_alpha=0.5, orig_dest_size=100, ax=None,
                bgcolor="white", node_color=node_color, node_size=node_size, edge_linewidth=1, edge_color='lightgray'
            )
```
